### Exercício 1
Carregue uma imagem colorida própria e exiba-a com correção de canais (BGR → RGB).

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread('foto_autoral.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
```
