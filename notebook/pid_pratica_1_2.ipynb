### Exercício 1
Carregue uma imagem colorida própria e exiba-a com correção de canais (BGR → RGB).

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_bgr = cv2.imread('foto_autoral.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
```

## Exercício 2
Imprima shape, size e dtype

```python
print("Shape:", img_rgb.shape)
print("Size:", img_rgb.size)
print("Dtype:", img_rgb.dtype)
```

## Exercício 3
Salva uma cópia da sua imagem original em um formato de compressão diferente

```python
cv2.imwrite("foto_autoral_convertida.jpg", cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR))
```

## Exercício 4
Acesse as coordenadas centrais da imagem e exiba os valores R, G e B desse
pixel específico

```python
h, w, _ = img_rgb.shape
pixel_central = img_rgb[h//2, w//2]
print("Pixel central (R, G, B):", pixel_central)
```

## Exercício 10

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, cmap="gray")
plt.axis("off")
```

## Exercício 13
Redimensione duas fotos para o mesmo tamanho e faça a mesclagem com
addWeighted (pesos 0.5/0.5)

```python
img2 = cv2.resize(img_rgb, (img_rgb.shape[1], img_rgb.shape[0]))
mesclada = cv2.addWeighted(img_rgb, 0.5, img2, 0.5, 0)

plt.imshow(mesclada)
plt.axis("off")
```

## Exercício 14
Implemente o negativo fotográfico da imagem em tons de cinzento

## Exercício 15
Crie uma máscara binária circular e use-a para isolar uma parte da sua foto
principal

```python
mask = np.zeros(img_gray.shape, dtype=np.uint8)
cv2.circle(mask, (w//2, h//2), min(w, h)//4, 255, -1)

resultado = cv2.bitwise_and(img_gray, img_gray, mask=mask)

plt.imshow(resultado, cmap="gray")
plt.axis("off")
```

## Exercício 16
Reduza a imagem original para 30% do tamanho inicial e exiba o novo shape

```python
resized = cv2.resize(img_rgb, None, fx=0.3, fy=0.3)
print("Novo shape:", resized.shape)

plt.imshow(resized)
plt.axis("off")
```

## Exercício 17
Calcule e imprima os valores de intensidade do histograma da sua imagem

```python
hist = cv2.calcHist([img_gray], [0], None, [256], [0,256])
print(hist.flatten())
```

## Exercício 18
Plote o histograma (gráfico) dos canais de cores separadamente no mesmo
gráfico

```python
colors = ('r','g','b')
for i, c in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color=c)

plt.title("Histograma RGB")
plt.show()
```

## Exercício 19
Aplique a Equalização de Histograma para melhorar o contraste de uma imagem
escura sua

```python
equalizada = cv2.equalizeHist(img_gray)

plt.imshow(equalizada, cmap="gray")
plt.axis("off")
```

## Exercício 20
Aplique um limiar (Threshold) binário simples para separar o objeto do fundo
